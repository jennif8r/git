#1. Arad -> Zerind

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt


file_path = '/content/test.csv'
distance_matrix = pd.read_csv(file_path, delimiter=';')

G = nx.from_pandas_adjacency(distance_matrix.set_index('Unnamed: 0'))


origem = 'arad'
destino = 'zerind'

if origem not in G.nodes():
    print("Nó de origem não encontrado no grafo.")
elif destino not in G.nodes():
    print("Nó de destino não encontrado no grafo.")
else:
    shortest_path = nx.dijkstra_path(G, source=origem, target=destino)
    distancia = nx.dijkstra_path_length(G, source=origem, target=destino)
    shortest_path_graph = G.subgraph(shortest_path)

    print(f"Caminho mais curto de {origem} até {destino}: {' -> '.join(shortest_path)}, Distância: {distancia}")

    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_size=500, node_color='skyblue', font_size=10, font_weight='bold')
    nx.draw_networkx_edges(G, pos, edgelist=shortest_path_graph.edges(), width=2, edge_color='red')
    plt.title(f"Caminho mais curto de {origem} até {destino}")
    print('')
    plt.show()
```
